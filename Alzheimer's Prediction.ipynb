```python
def remove_duplicate_columns(dframe):
    new_dataFrame = pd.DataFrame(index=dframe.index)
    for i,col in enumerate(dframe.columns):
        if not col in new_dataFrame.columns:
            try:
                new_dataFrame[col] = np.array(dframe[col])[:]
            except:
                new_dataFrame[col] = np.array(dframe[col])[:,0]
                
    return new_dataFrame
alexa_new = remove_duplicate_columns(alexa_new)
v1_new = remove_duplicate_columns(v1_new)
```

```python
labels = v1_new[['DiseaseStatus']].copy()
labels = list(labels['DiseaseStatus'] == 'Control') #control is class 1, alzheimers class 0
data_columns =v1_new.columns

from sklearn import preprocessing
import matplotlib
import matplotlib.pyplot as plt

features = v1_new.drop(['DiseaseStatus','PatientID'], axis=1)
features = preprocessing.scale(features) #standard normal distribution

plt.hist(features.flatten(), bins=80) #histogram for features after standard normalization
```

```python
from sklearn.model_selection import train_test_split

X_train, X_validation, y_train, y_validation = train_test_split(features, labels, test_size=0.3, stratify=labels) #validation = to check for overfitting; if overfit = regularize #so now our working data is 305 points. we had original 339 - 33.9 (1% of 339) = 305
features = X_train
labels = y_train
#print(X_train)
```

```python
labels2 = alexa_new[['DiseaseStatus']].copy()
labels2 = list(labels2['DiseaseStatus'] == 'Control') #control is class 1, alzheimers class 0
data_columns2 =alexa_new.columns

from sklearn import preprocessing
import matplotlib
import matplotlib.pyplot as plt

features2= alexa_new.drop(['DiseaseStatus','PatientID'], axis=1)
features2= preprocessing.scale(features2) #standard normal distribution

plt.hist(features2.flatten(), bins=80) #histogram for features after standard normalization
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_validation)
knn_r = round(accuracy_score(y_pred, y_validation) * 100, 2)
print(knn_r)
```

```python
y_pred2=knn.predict(features2)
knn_r = round(accuracy_score(y_pred2,labels2) * 100, 2)
print(knn_r)
```
